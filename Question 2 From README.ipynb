```
%reset
import math
import numpy as np
import tensorflow as tf
import tensorflow.keras as keras
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras.datasets import imdb

(training_data, training_targets), (testing_data, testing_targets) = imdb.load_data(num_words=10000)

def vectorize(sequences, dimension = 10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1
    return results

train_x = vectorize(training_data)
train_x = np.insert(train_x, 0, 1, axis=1);

test_x = vectorize(testing_data)
test_x = np.insert(test_x, 0, 1, axis=1);

train_y = training_targets

test_y = testing_targets

```

```
def sigmoid(x):
  return 1 / (1 + math.exp(-x))



```

```
N=25000
d=10000
eta=5
rng = np.random.RandomState(1)
w = rng.randn(1, d+1)/100;
dwacc=np.zeros(w.shape)

for epoch in range(400):
  numcorr=0
  dw=np.zeros(w.shape)
  for t in range(N):
    r=train_y[t]
    x=train_x[t,:]
    y=sigmoid(np.dot(x,w.T))
    if (y>0.5 and r==1) or (y<=0.5 and r==0) :
      numcorr=numcorr+1
    delta=r-y
    dw=dw+x*delta
  dwacc = 0*dwacc + 1*dw/N
  w=w*1+eta*dwacc;
  print(numcorr/N, np.sum(w**2))  
    
```
