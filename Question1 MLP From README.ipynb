```
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
iris = datasets.load_iris()
train_x = np.insert(iris.data, 0, 1, axis=1);
train_y = iris.target
train_y[:50]=0
train_y[50:100]=1
train_y[100:150]=0
#train_y=train_y-1
#train_x=train_x[50:,:]
print(train_y)
```

# New Section

```
N=150
H=50
d=4
etav=0.1
etaw=0.5
rng = np.random.RandomState(1)
v = rng.randn(1, H+1)
w = rng.randn(H, d+1)
z = np.ones(H+1)
print(z.shape)
print(z)
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

for epoch in range(100):
  numcorr=0
  dv=np.zeros(v.shape)
  dw=np.zeros(w.shape)
  for t in range(N):
    r=train_y[t]
    x=train_x[t,:]
    for h in range(1,H+1):
      z[h]=sigmoid(np.dot(w[h-1,:],x))
    y=sigmoid(np.dot(v,z))
    if (y>0.5 and r==1) or (y<=0.5 and r==0) :
      numcorr=numcorr+1
    delta=r-y
    dv=dv+z*delta
    for h in range(1,H+1):
      for j in range(d+1):
        dw[h-1,j]=dw[h-1,j] + delta * v[0,h] * z[h] * (1-z[h]) * x[j]
  v=v+etav*dv/N
  w=w+etaw*dw/N
  print(numcorr)  
```
